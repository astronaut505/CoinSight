```python
import yfinance as yf

# Get the data for Bitcoin
data = yf.download('BTC-USD')

# Display the data
print(data)
```

```python
data.plot(y='Close')
```

```python
import pandas as pd
import numpy as np

# Calculate the 12-day Rate of Change (ROC)
data['momentum_12'] = (data['Close'] - data['Close'].shift(12)) / data['Close'].shift(12) * 100

# Calculate the MACD indicator
exp12 = data['Close'].ewm(span=12, adjust=False).mean()
exp26 = data['Close'].ewm(span=26, adjust=False).mean()
macd_line = exp12 - exp26
signal_line = macd_line.ewm(span=9, adjust=False).mean()
data['macd'] = macd_line
data['macd_signal'] = signal_line

# Calculate the RSI indicator
delta = data['Close'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
ema_up = up.ewm(com=13, adjust=False).mean()
ema_down = down.ewm(com=13, adjust=False).mean()
rs = ema_up/ema_down
data['rsi'] = 100 - (100/(1 + rs))

# Calculate the Stochastic RSI indicator
n = 14
low_14 = data['Close'].rolling(n).min()
high_14 = data['Close'].rolling(n).max()
data['stoch_rsi'] = (data['Close'] - low_14) / (high_14 - low_14)

# Calculate the Bollinger Bands
data['bb_middle'] = data['Close'].rolling(window=20).mean()
data['bb_upper'] = data['bb_middle'] + 2*data['Close'].rolling(window=20).std()

# Calculate the Williams Percentage Range (14 days)
high14 = data['High'].rolling(14).max()
low14 = data['Low'].rolling(14).min()
data['williams_pct'] = (high14 - data['Close']) / (high14 - low14) * -100

# Display the data with the calculated indicators
print(data)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(7, figsize=(14, 14))

# Plot Close price
axs[0].plot(data['Close'], label='Close Price', color='blue')
axs[0].set_title('Close Price')
axs[0].legend()

# Plot ROC
axs[1].plot(data['momentum_12'], label='12-day ROC', color='orange')
axs[1].set_title('Rate of Change')
axs[1].legend()

# Plot MACD
axs[2].plot(data['macd'], label='MACD', color='red')
axs[2].plot(data['macd_signal'], label='Signal Line', color='green')
axs[2].set_title('MACD')
axs[2].legend()

# Plot RSI
axs[3].plot(data['rsi'], label='RSI', color='purple')
axs[3].set_title('Relative Strength Index')
axs[3].legend()

# Plot Stochastic RSI
axs[4].plot(data['stoch_rsi'], label='Stochastic RSI', color='brown')
axs[4].set_title('Stochastic RSI')
axs[4].legend()

# Plot Bollinger Bands
axs[5].plot(data['Close'], label='Close Price', color='blue')
axs[5].plot(data['bb_middle'], label='Middle Band', color='red')
axs[5].plot(data['bb_upper'], label='Upper Band', color='green')
axs[5].set_title('Bollinger Bands')
axs[5].legend()

# Plot Williams %R
axs[6].plot(data['williams_pct'], label='Williams %R', color='black')
axs[6].set_title('Williams %R')
axs[6].legend()

# Display the plot
plt.tight_layout()
plt.show()
```
